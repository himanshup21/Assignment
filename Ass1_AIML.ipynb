```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 9.5367431640625*10**-8
```

```python
y = [dt*i for i in range(4000000)] 
```

```python
x = [0.1*i/3600 for i in range(4000000)]
```

```python
print((x[50*36000],y[50*36000]))
```

```python
print((x[100*36000],y[100*36000]))
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.plot(x[50*36000], y[50*36000], marker="o", markersize=20, markerfacecolor="green")
plt.plot(x[100*36000], y[100*36000], marker="o", markersize=20, markerfacecolor="red")
plt.xlabel("actual time in hrs",fontsize=20)
plt.ylabel("error in secs",fontsize=20)
plt.savefig('error.png')
```

```python
y1 = [i*3000 for i in y]
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(x,y1)
plt.plot(x[1747626], y1[1747626], marker="o", markersize=20, markerfacecolor="red")
plt.xlabel("actual time in hrs",fontsize=20)
plt.ylabel("Xerr in meters",fontsize=20)
plt.savefig('error1.png')
```
